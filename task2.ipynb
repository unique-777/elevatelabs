```python
#seperating numerical and actegorical columns
numerical = data.select_dtypes(["int","float"]).columns
categorical = data.select_dtypes(["object","category"]).columns
```

```python
#replacing the median values in place of missing and null values
imputer = SimpleImputer(strategy='median')
data[numerical] = imputer.fit_transform(data[numerical])
```

```python
data.hist(bins=20, figsize=(12, 8))
plt.tight_layout()#automatically adjust subplot parameters so that your plots don't overlap and everything fits nicely within the figure
plt.show()
for col in numerical:
    plt.figure()
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#numerical features correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(data[numerical].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
